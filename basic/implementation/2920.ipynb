##내 코드




```
numlist=input().split()

def isEight(numlist):
  if len(numlist)!=8:
    return False
  return True

def ascending(numlist):
  if numlist==['1', '2', '3', '4', '5', '6', '7', '8']:
    return True
  else:
    return False



  # i=0
  # for num in numlist:
  #   i+=1
  #   if int(num)!=i:
  #     return False
  # return True

def descending(numlist):
  if numlist==['8', '7', '6', '5', '4', '3', '2', '1']:
    return True
  else:
    return False

  # i=9
  # for num in numlist:
  #   i-=1
  #   if int(num)!=i:
  #     return False
  # return True

def mixed(numlist):
  if isEight(numlist)==False:
    return False
  for num in numlist:
    if int(num)<1 or int(num)>8:
      return False
  return True


if ascending(numlist)==True:
  print("ascending")
elif descending(numlist)==True:
  print("descending")
elif mixed(numlist)==True:
  print("mixed")
```

##풀이코드

```
numlist=list(map(int,input().split()))

ascending=True
descending=True


for i in range(1,8):
  if numlist[i]>=numlist[i-1]:
    descending=False
  elif numlist[i]<=numlist[i-1]:
    ascending=False

if ascending==True:
  print("ascending")
elif descending==True:
  print("descending")
else:
  print("mixed")

```
