```
N=int(input())

result=-1
max_kilo5=N//5

for num_kilo5 in range(max_kilo5,-1,-1):
  p_bag=num_kilo5
  remain=N-num_kilo5*5
  if remain%3==0:
    p_bag+=remain//3
    result=p_bag
    break

print(result)
```

```
N=int(input())
bag=0
while (N>=3):
  if N%5==0:
    bag+=N//5
    N=0
    break
  else:
    N=N-3
    bag+=1

if N!=0:
  print(-1)
else:
  print(bag)



```
