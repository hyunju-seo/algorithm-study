```
N, M=map(int,input().split())
numList=list(map(int,input().split()))
#N 3이상 100 이하// M 10 이상 30만 이
numList.sort()
sumList=[]


for i in range(N-1,-1,-1):
  for k in range(i-1,-1,-1):
    for j in range(k-1,-1,-1):
      # print(i,k,j)
      # print(numList[N-i-1],numList[i-k-1],numList[k-j-1])
      sum=numList[i]+numList[k]+numList[j]
      if sum==M:
        sumList.append(sum)
        break
      if sum<M:
        sumList.append(sum)
print(max(sumList))


```

##풀이코드
파이썬은 일초에 2천만 정도의 계산은 수행하는데 이 문제에서 3중 for 문을 이용하면 최대 100만 정도의 계산을 수행하므로 충분

```
n,m=map(int,input().split())
data=list(map(int,input().split()))

result=0

for i in range(n):
  for j in range(i+1,n):
    for k in range(j+1,n):
      sum=data[i]+data[j]+data[k]
      if sum<=m:
        result=max(result,sum)  # 항상 max를 비교
print(result)

```
